```python
def calcula():
    operacao = input('''Selecione a operação desejada:
+ para adição
- para subtração
* para multiplicação
/ para divisão
X para sair''')

    if operacao.upper() == 'X':
        print("Fim do programa")
        
    else:
        
        n1 = int(input('Número 1: '))
        n2 = int(input('Número 2: '))
    
                
        if operacao == '+':
            print('{} + {} = '.format(n1, n2))
            print(n1 + n2)

        elif operacao == '-':
            print('{} - {} = '.format(n1, n2))
            print(n1 - n2)

        elif operacao == '*':
            print('{} * {} = '.format(n1, n2))
            print(n1 * n2)

        elif operacao == '/':

            if (n1 == 0) or (n2 == 0):
                print('Operação inválida')
            else:
                print('{} / {} = '.format(n1, n2))
                print(n1 / n2)
            
        else:
            print('Operação inválida')

        repetir()

def repetir():
    calc_repetir = input('''
Você gostaria de reiniciar o programa?

Digite S para Sim e N para Não:''')

    if calc_repetir.upper() == 'S':
        calcula()
    elif calc_repetir.upper() == 'N':
        print('Fim do programa.')
    else:
        repetir()

calcula()
```
